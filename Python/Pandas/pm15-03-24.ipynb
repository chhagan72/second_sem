```python
# sorting
empData={
    'EmpId':[101,102,103,104,105,106],
    'Ename':['pm','ram','joy','raj','vaib','chhagan' ],
    'salary':[10000,20000,40000,25000,15000,60000]
}
df=pd.DataFrame(empData)
print(df,"\n")
a=df.sort_values('salary')
print(a)
```

```python
# sorting on multiple columns    --> ascending b default is true for decending use false
b=df.sort_values(by=['Ename','salary'],ascending=True)
print(b)
```

```python
# sort index method
c=df.sort_index(ascending=False)
print(c)
```

```python
# rename index
d=df.rename(index={0:"1st",1:"2nd",2:"3rd",3:"4th",4:"5th",5:"6th"})
print(d)
```

```python
# delete index
e=df.reset_index().drop(index=[2,3])
print(e)
```

```python
data={
    'year':['2018','2019','2018','2019','2018','2019'],
    'Team':['MI','MI','RCB','RCB','CSK','CSK'],
    'Runs':[2500,2650,2200,2400,2300,2700]
}
df=pd.DataFrame(data)
pv=pd.pivot(df,index='year',columns='Team',values='Runs')
print(df,"\n")
print(pv)
```

```python
# pivot table
data={
    'Date':['1-1-2019','1-1-2019','1-2-2019','1-2-2019','1-3-2019','1-3-2019'],
    'city':['Delhi','Delhi','Mumbai','Mumbai','Chennai','Chennai'],
    'Temp':[28,30,22,24,32,34],
    'Humidity':[60,55,80,70,90,85]
}
df=pd.DataFrame(data)
print(df,'\n')
pv=pd.pivot_table(df,index='city',values='Temp')
print(pv)
```

```python
pv=pd.pivot_table(df,index='city',values='Temp',aggfunc=max)
print(pv)
```
